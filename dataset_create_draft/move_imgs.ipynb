```python
dirs = [dir_ for dir_ in os.listdir('lgg-mri-segmentation/kaggle_3m/') if len(dir_) > 20]
cnt = 0
for dir_ in dirs:
    cnt += 1
    print(dir_)
    path = 'lgg-mri-segmentation/kaggle_3m/' + dir_
    imgs = [img for img in os.listdir(path) if 'mask' not in img]
    if cnt < 101:
        for img in imgs:
            shutil.move('lgg-mri-segmentation/kaggle_3m/' + dir_ + '/' + img, 'trainA')
    else:
        for img in imgs:
            shutil.move('lgg-mri-segmentation/kaggle_3m/' + dir_ + '/' + img, 'testA')
```

```python
dirs = os.listdir('computed-tomography-images-for-intracranial-hemorrhage-detection-and-segmentation-1.0.0/Patients_CT/')
cnt = 0
for dir_ in dirs:
    cnt += 1
    print(dir_)
    path = 'computed-tomography-images-for-intracranial-hemorrhage-detection-and-segmentation-1.0.0/Patients_CT/' + dir_ + '/brain/'
    imgs = [img for img in os.listdir(path) if 'HGE' not in img]
    src = 'computed-tomography-images-for-intracranial-hemorrhage-detection-and-segmentation-1.0.0/Patients_CT/' + dir_ + '/brain/' 
    if cnt < 74:
        for img in imgs:           
            os.rename(src + img, src + dir_ + '-' + img)
            shutil.move(src + dir_ + '-' + img, 'trainB')
    else:
        for img in imgs:
            os.rename(src + img, src + dir_ + '-' + img)
            shutil.move(src + dir_ + '-' + img, 'testB')
```
